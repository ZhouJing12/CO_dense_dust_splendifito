```python
import emcee
import numpy as np
import matplotlib.pyplot as plt
```

```python
sampler=emcee.backends.HDFBackend('fit_results/NGC6240_sample.h5',read_only=True)
```

```python
samples=sampler.get_chain(discard=7000)
```

```python
plt.figure(figsize=[4.5,4])
plt.yticks([])
plt.hist(samples[:,:,0].reshape(13000*40),bins=100,range=[1,3],histtype='step')
plt.hist(samples[:,:,4].reshape(13000*40),bins=100,range=[1,3],histtype='step')
plt.hist(samples[:,:,8].reshape(13000*40),bins=100,range=[1,3],histtype='step')
plt.xlabel(r'log $T_{\rm kin}$ (K)')
plt.savefig('/home/zj/Downloads/Tkin_pdf.pdf')
```

```python
plt.figure(figsize=[4,5])
xx=np.linspace(-100,100,10000)
xx_t1=xx[(xx<7)*(xx>1)]
xx_t2=xx[(xx<9)*(xx>3)]
xx_t3=xx[(xx<11)*(xx>5)]
plt.plot(xx_t1,np.exp(-(xx_t1-4)**2),'purple')
plt.plot(xx_t1,np.exp(-(xx_t1-4)**2)+1.3,color='purple',linestyle='dashed')
plt.plot(xx_t2,np.exp(-(xx_t2-6)**2)+1.3,color='purple')
plt.plot(xx_t1,np.exp(-(xx_t1-4)**2)+2.6,color='purple',linestyle='dashed')
plt.plot(xx_t2,np.exp(-(xx_t2-6)**2)+2.6,color='purple',linestyle='dashed')
plt.plot(xx_t3,np.exp(-(xx_t3-8)**2)+2.6,color='purple')
plt.arrow(4,1.1,0,2.2,head_length=0.2,width=0.3,fc='green',ec='green',alpha=0.3)
plt.arrow(6,2.4,0,0.9,head_length=0.2,width=0.3,fc='green',ec='green',alpha=0.3)
plt.text(10,3.5,'LVG',fontsize=15)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
plt.xlim(0,12.5)
plt.ylim(0,4)
plt.yticks([])
plt.xticks([])
plt.xlabel('velocity')
plt.ylabel('S')
plt.savefig('/home/zj/Downloads/LVG_show.pdf')
```
